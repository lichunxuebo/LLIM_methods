```python
# coding=utf-8
import cv2
import numpy as np
# cv2.imshow(img) 使内核崩溃
# 当使用的是 google colab，则可以使用此解决方案：
# from google.colab.patches import cv2_imshow
# cv2_imshow(img)
from google.colab.patches import cv2_imshow

# 读取图片
# cv2.imread(地址,flag)
# flag = -1,   8位深度，原通道
# flag = 0，   8位深度，1通道
# flag = 1，   8位深度，3通道
# flag = 2，   原深度， 1通道
# flag = 3，   原深度， 3通道
# flag = 4，   8位深度，3通道
image = cv2.imread("Jimmy.jpg",3)
print(image.shape)
# 高为image.shape[0]，宽为image.shape[1]
print(image.shape[0],image.shape[1])
# 创建图片【空的查找表】
# grayImg = cv.CreateImage(cv.GetSize(image), image.depth, 1)
grayImg = np.zeros(image.shape,np.uint8)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        grayImg[i,j] = 0.3*image[i,j][0]+0.59*image[i,j][1]+0.11*image[i,j][2]
cv2_imshow(image)
cv2_imshow(grayImg)
cv2.imwrite("Jimmy03.jpg", grayImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
